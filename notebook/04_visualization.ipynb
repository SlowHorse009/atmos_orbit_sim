```python
import pandas as pd
import plotly.graph_objects as go

# 1. 读取我们刚存好的完全体数据
df = pd.read_csv('integrated_mission_data.csv')

# 2. 创建 3D 交互式地球画布
fig = go.Figure()

# 3. 绘制卫星本体轨迹 (蓝色实线)
fig.add_trace(go.Scattergeo(
    lon=df['Sat_Lon'],
    lat=df['Sat_Lat'],
    mode='lines+markers',
    line=dict(width=3, color='royalblue'),
    marker=dict(size=4, color='royalblue'),
    name='卫星真实飞行轨迹',
    hoverinfo='text',
    text=df.apply(lambda row: f"时间: {row['UTC_Time']}<br>高度: {row['Sat_Alt_km']} km", axis=1)
))

# 4. 绘制大气探测切点轨迹 (热力图映射温度)
fig.add_trace(go.Scattergeo(
    lon=df['Target_Lon'],
    lat=df['Target_Lat'],
    mode='markers',
    marker=dict(
        size=6,
        color=df['Target_Temp_K'],
        colorscale='Inferno',
        colorbar_title="切点温度 (K)",
        showscale=True
    ),
    name='大气切点跟瞄轨迹',
    hoverinfo='text',
    text=df.apply(lambda row: f"俯仰角: {row['Pitch_Deg']}°<br>切点高度: {row['Target_Alt_km']} km<br>温度: {row['Target_Temp_K']} K", axis=1)
))

# 5. 配置 3D 地球外观 (上帝视角)
fig.update_layout(
    title='全链路临边探测交互式 3D 仿真地图 (Pro版)',
    geo=dict(
        projection_type='orthographic', # 开启 3D 地球模式
        showland=True,
        landcolor="rgb(212, 212, 212)",
        showocean=True,
        oceancolor="rgb(10, 30, 60)",
        showlakes=False,
        showcountries=True,
        countrycolor="rgba(255, 255, 255, 0.3)"
    ),
    margin=dict(l=0, r=0, t=50, b=0)
)

# 6. 渲染出图！
fig.show()
```
