```python
import math
import sys
# 假设你已经建好了上文的 src 文件夹和两个脚本
sys.path.append('../src') 
from geometry import calculate_depression_angle
from sensor import GSENSE4040

# 固定轨道高度和目标高度
H_SAT_KM = 500.0         
H_TARGET_KM = 20.0       

# 我们测试三组不同的镜头焦距 (毫米)
test_focal_lengths = [50.0, 100.0, 200.0]

print(f"【固定条件】轨道高度: {H_SAT_KM} km | 最低切点: {H_TARGET_KM} km")
print("-" * 50)

for f in test_focal_lengths:
    sensor = GSENSE4040(focal_length_mm=f)
    alpha_deg, distance_km = calculate_depression_angle(H_SAT_KM, H_TARGET_KM)
    
    # 空间分辨率
    ifov_rad = sensor.get_ifov()
    resolution_m = (distance_km * 1000) * ifov_rad
    
    # 垂直覆盖范围
    total_fov_rad = sensor.SENSOR_SIZE_MM / f
    vertical_swath_km = distance_km * total_fov_rad
    
    # 判断能否覆盖 480 km (20km - 500km)
    can_cover = "✅可以" if vertical_swath_km >= 480 else "❌不能"
    
    print(f"▶ 焦距: {f} mm")
    print(f"  - 单像元分辨率: {resolution_m:.1f} 米")
    print(f"  - 垂直总视野: {vertical_swath_km:.1f} km -> 单次拍全 20-500km? {can_cover}")
    print("-" * 50)
```
