```python
import pandas as pd
import numpy as np

# 1. 像前端一样读取文件 (引擎已经帮你落盘了)
file_path = "output/sim_result_tgt400km.parquet"
df = pd.read_parquet(file_path)

# 2. 基础信息全览
print(f"✅ 成功加载推演数据，总帧数: {len(df)}")
print("-" * 50)
print(df.info()) 
print("-" * 50)

# 3. 针对性检验：看看你刚才关心的“完美条件下的残差”到底有多大！
target_expected = 400.0  # 预期目标高度 400km
actual_alts = df['Tgt_Alt_km'].values
residuals_meters = (actual_alts - target_expected) * 1000  # 转换为米

print(f"🎯 靶心高度残差分析 (WGS84 截断误差):")
print(f"最大偏差: {np.max(np.abs(residuals_meters)):.2f} 米")
print(f"平均偏差: {np.mean(residuals_meters):.2f} 米")

# 4. 瞄一眼前 5 行核心数据
display(df[['T_sec', 'Cmd_Pitch_deg', 'Tgt_Alt_km', 'Illum_Sat']].head())
```
